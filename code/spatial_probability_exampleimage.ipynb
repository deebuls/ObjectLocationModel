```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({ 'A' : pd.Categorical(["kitchen","table","kitchen","Bedroom", "Dish washer",
                                         "table","table","table","table","kitchen","kitchen"
                                         ,"kitchen","kitchen","kitchen","kitchen"])})

df['A'].value_counts()

print (df['A'].count())
```

```python
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
ax = (df['A'].value_counts()/df['A'].count()).plot(kind='bar', rot=0)
ax.set_title("Probability of an object at different locations")
ax.set_xlabel("Locations")
ax.set_ylabel("Probability")
```
